# Function 5: 4D Chemical Process Yield

*Analogy: **Typically unimodal; single peak.** We **maximize** \(y\) (e.g. process yield or conversion efficiency).*


### Characteristics


* **4D input**: \((x_1, x_2, x_3, x_4)\) — e.g. temperature, pressure, catalyst, flow-rate parameters in \([0,1]^4\).


* **Objective**: Maximize \(y\). Initial data: 20 points, shape (20, 4).


* **Strategy**: Same Bayesian optimisation workflow — GP surrogate, EI/PI/UCB acquisition — in 4D; visualisation uses 2D pairwise slices (6 pairs).

## 3. Suggest next point (Bayesian optimisation)

Fit three GP kernels, select best by log-marginal-likelihood; compute EI, PI, UCB; choose next query (solo or ensemble).

### 3.1 Surrogate: GP kernels

- **Gaussian Process (GP)** — We compare three kernel choices; each trades off smoothness, noise handling, and flexibility.
- **RBF** (squared exponential) — Infinitely smooth; good default.
- **Matérn (ν=1.5)** — Once-differentiable; more flexible for rough surfaces.
- **RBF + WhiteKernel** — RBF with explicit observation noise; use when the oracle is stochastic.

Best kernel is chosen by log-marginal-likelihood (or set `GP_KERNEL` manually).

```python
_sampler = Sobol() if CANDIDATE_SAMPLING_METHOD == "sobol" else Lhs()
candidate_pts = np.array(_sampler.generate([(0.0, 1.0)] * 4, n_cand))

_gp_optimizer = None if not OPTIMIZE_KERNEL else 'fmin_l_bfgs_b'
_gp_n_restarts = N_RESTARTS_KERNEL if OPTIMIZE_KERNEL else 0
kernel_RBF       = ConstantKernel(CONSTANT_KERNEL_SCALE, constant_value_bounds=CONSTANT_SCALE_BOUNDS) * RBF(length_scale=LENGTH_SCALE, length_scale_bounds=LENGTH_SCALE_BOUNDS)
kernel_Matern    = ConstantKernel(CONSTANT_KERNEL_SCALE, constant_value_bounds=CONSTANT_SCALE_BOUNDS) * Matern(length_scale=LENGTH_SCALE, nu=MATERN_NU, length_scale_bounds=LENGTH_SCALE_BOUNDS)
kernel_RBF_noise = ConstantKernel(CONSTANT_KERNEL_SCALE, constant_value_bounds=CONSTANT_SCALE_BOUNDS) * RBF(length_scale=LENGTH_SCALE, length_scale_bounds=LENGTH_SCALE_BOUNDS) + WhiteKernel(noise_level=WHITE_NOISE_LEVEL, noise_level_bounds=WHITE_NOISE_BOUNDS)
gp_RBF       = GaussianProcessRegressor(kernel=kernel_RBF,       alpha=GP_ALPHA, optimizer=_gp_optimizer, n_restarts_optimizer=_gp_n_restarts, normalize_y=True)
gp_Matern    = GaussianProcessRegressor(kernel=kernel_Matern,    alpha=GP_ALPHA, optimizer=_gp_optimizer, n_restarts_optimizer=_gp_n_restarts, normalize_y=True)
gp_RBF_noise = GaussianProcessRegressor(kernel=kernel_RBF_noise, alpha=GP_ALPHA, optimizer=_gp_optimizer, n_restarts_optimizer=_gp_n_restarts, normalize_y=True)
gp_RBF.fit(X, y); gp_Matern.fit(X, y); gp_RBF_noise.fit(X, y)
print(f"Kernel hyperparameters: {'optimized (MLE)' if OPTIMIZE_KERNEL else 'fixed (no optimization)'}")
mu_cand_RBF, sigma_cand_RBF = gp_RBF.predict(candidate_pts, return_std=True)
mu_cand_Matern, sigma_cand_Matern = gp_Matern.predict(candidate_pts, return_std=True)
mu_cand_RBF_noise, sigma_cand_RBF_noise = gp_RBF_noise.predict(candidate_pts, return_std=True)
gp_neg_RBF = copy.deepcopy(gp_RBF); gp_neg_Matern = copy.deepcopy(gp_Matern); gp_neg_RBF_noise = copy.deepcopy(gp_RBF_noise)
gp_neg_RBF.fit(X, -y); gp_neg_Matern.fit(X, -y); gp_neg_RBF_noise.fit(X, -y)
print("GP RBF:", gp_RBF.kernel_)
```

```python
gps = [
    (gp_RBF,       gp_neg_RBF,       "RBF"),
    (gp_Matern,    gp_neg_Matern,    "Matérn (ν=1.5)"),
    (gp_RBF_noise, gp_neg_RBF_noise, "RBF + WhiteKernel"),
]
CANONICAL_NAMES = ["RBF", "Matérn (ν=1.5)", "RBF + WhiteKernel"]
KERNEL_ALIASES = {
    "rbf": "RBF", "matern": "Matérn (ν=1.5)", "matérn (ν=1.5)": "Matérn (ν=1.5)",
    "matern (v=1.5)": "Matérn (ν=1.5)", "rbf + whitekernel": "RBF + WhiteKernel",
    "rbf+whitekernel": "RBF + WhiteKernel", "rbf + white": "RBF + WhiteKernel", "white": "RBF + WhiteKernel",
}

if GP_KERNEL is None or str(GP_KERNEL).strip().lower() in ("none", "auto", ""):
    lml = {name: gp.log_marginal_likelihood_value_ for gp, _, name in gps}
    best_name = max(lml, key=lml.get)
    print("LML scores:", {k: f"{v:.2f}" for k, v in lml.items()})
    print(f"Auto-selected kernel: {best_name} (highest LML)")
else:
    key = str(GP_KERNEL).strip().lower()
    best_name = KERNEL_ALIASES.get(key, GP_KERNEL)
    if best_name not in CANONICAL_NAMES:
        raise ValueError(f"Unknown GP_KERNEL='{GP_KERNEL}'. Use one of {CANONICAL_NAMES}")
    print(f"Manual kernel override: {best_name}")

is_rbf    = best_name == "RBF"
is_matern = best_name == "Matérn (ν=1.5)"
is_rbf_w  = best_name == "RBF + WhiteKernel"

best_gp     = gp_RBF if is_rbf else (gp_Matern if is_matern else gp_RBF_noise)
best_gp_neg = gp_neg_RBF if is_rbf else (gp_neg_Matern if is_matern else gp_neg_RBF_noise)
print(f"OPTIMIZE_KERNEL = {OPTIMIZE_KERNEL}")
```

```python
n_slice = 50
ug = np.linspace(0, 1, n_slice)
Ug, Vg = np.meshgrid(ug, ug)
med = [np.median(X[:, k]) for k in range(4)]
slices_info = []
for ia, ib, la, lb in pairs:
    other = [k for k in range(4) if k not in (ia, ib)]
    slice_pts = np.zeros((Ug.size, 4))
    slice_pts[:, ia] = Ug.ravel()
    slice_pts[:, ib] = Vg.ravel()
    for k in other:
        slice_pts[:, k] = med[k]
    held = ", ".join(f"x_{k+1}={med[k]:.2f}" for k in other)
    slices_info.append((slice_pts, ia, ib, la, lb, held))

fig, axes = plt.subplots(6, 2, figsize=(10, 24))
for row, (slice_pts, ia, ib, la, lb, held) in enumerate(slices_info):
    for gp_obj, _, gp_name in gps:
        if gp_name != best_name:
            continue
        mu_s, sig_s = gp_obj.predict(slice_pts, return_std=True)
        mu_s = mu_s.reshape(Ug.shape); sig_s = sig_s.reshape(Ug.shape)
    ax_mu, ax_sig = axes[row, 0], axes[row, 1]
    _mu_min, _mu_max = np.nanmin(mu_s), np.nanmax(mu_s)
    if _mu_max <= _mu_min: _mu_max = _mu_min + 1e-10
    lev_mu = np.linspace(_mu_min, _mu_max, CONTOURF_LEVELS)
    cf1 = ax_mu.contourf(Ug, Vg, mu_s, levels=lev_mu, cmap="viridis")
    ax_mu.contour(Ug, Vg, mu_s, levels=np.linspace(_mu_min, _mu_max, CONTOUR_LINE_LEVELS),
                  colors="k", linewidths=CONTOUR_LINEWIDTH, alpha=CONTOUR_ALPHA, linestyles=CONTOUR_LINESTYLE)
    fig.colorbar(cf1, ax=ax_mu).set_label("μ(x)", fontsize=10)
    ax_mu.scatter(X[:, ia], X[:, ib], c="red", s=60, edgecolors="k", zorder=2, linewidths=1.5)
    ax_mu.set_xlabel(la, fontsize=10); ax_mu.set_ylabel(lb, fontsize=10)
    ax_mu.set_title(f"{best_name} — mean ({held})", fontsize=10)
    ax_mu.set_aspect("equal"); ax_mu.set_xlim(0, 1); ax_mu.set_ylim(0, 1)

    _sig_min, _sig_max = np.nanmin(sig_s), np.nanmax(sig_s)
    if _sig_max <= _sig_min: _sig_max = _sig_min + 1e-10
    lev_sig = np.linspace(_sig_min, _sig_max, CONTOURF_LEVELS)
    cf2 = ax_sig.contourf(Ug, Vg, sig_s, levels=lev_sig, cmap="plasma")
    ax_sig.contour(Ug, Vg, sig_s, levels=np.linspace(_sig_min, _sig_max, CONTOUR_LINE_LEVELS),
                   colors="b", linewidths=CONTOUR_LINEWIDTH, alpha=CONTOUR_ALPHA, linestyles=CONTOUR_LINESTYLE)
    fig.colorbar(cf2, ax=ax_sig).set_label("σ(x)", fontsize=10)
    ax_sig.scatter(X[:, ia], X[:, ib], c="red", s=60, edgecolors="k", zorder=2, linewidths=1.5)
    ax_sig.set_xlabel(la, fontsize=10); ax_sig.set_ylabel(lb, fontsize=10)
    ax_sig.set_title(f"{best_name} — std ({held})", fontsize=10)
    ax_sig.set_aspect("equal"); ax_sig.set_xlim(0, 1); ax_sig.set_ylim(0, 1)

fig.suptitle(f"GP surrogate: posterior mean (μ) and uncertainty (σ)  [{best_name}]  |  warping: {WARP_LABEL}", fontsize=13, y=1.005)
fig.tight_layout()
if IF_EXPORT_PLOT:
    out_dir = PLOT_EXPORT_DIR; out_dir.mkdir(parents=True, exist_ok=True)
    fig.savefig(out_dir / "function_5_gp_slices.png", dpi=150, bbox_inches="tight")
if IF_SHOW_PLOT:
    plt.show()
```

### 3.2 Acquisition functions* Expected Improvement (EI)* Upper Confidence Bound (UCB)* Probability of Improvement (PI)

```python
y_best = best_y

EI_RBF  = gaussian_ei(candidate_pts, gp_neg_RBF, y_opt=-y_best, xi=XI_EI_PI)
PI_RBF  = gaussian_pi(candidate_pts, gp_neg_RBF, y_opt=-y_best, xi=XI_EI_PI)
LCB_RBF = gaussian_lcb(candidate_pts, gp_neg_RBF, kappa=KAPPA_UCB)
EI_Matern  = gaussian_ei(candidate_pts, gp_neg_Matern, y_opt=-y_best, xi=XI_EI_PI)
PI_Matern  = gaussian_pi(candidate_pts, gp_neg_Matern, y_opt=-y_best, xi=XI_EI_PI)
LCB_Matern = gaussian_lcb(candidate_pts, gp_neg_Matern, kappa=KAPPA_UCB)

EI_RBF_noise  = gaussian_ei(candidate_pts, gp_neg_RBF_noise, y_opt=-y_best, xi=XI_EI_PI)
PI_RBF_noise  = gaussian_pi(candidate_pts, gp_neg_RBF_noise, y_opt=-y_best, xi=XI_EI_PI)
LCB_RBF_noise = gaussian_lcb(candidate_pts, gp_neg_RBF_noise, kappa=KAPPA_UCB)

# Mask candidates too close to existing observations (avoid re-evaluating)
min_dist_cand = np.min(np.sqrt(((candidate_pts[:, None, :] - X[None, :, :]) ** 2).sum(axis=2)), axis=1)
next_x_high_dist = np.clip(np.asarray(candidate_pts[np.argmax(min_dist_cand)]).ravel(), 0.0, 0.999999)
too_close = min_dist_cand < MIN_DIST_THRESHOLD
try: BOUNDARY_MARGIN
except NameError: BOUNDARY_MARGIN = 0
# Optional: mask candidates near domain boundaries (low-d only; set BOUNDARY_MARGIN=0 for high-d)
near_boundary = np.any((candidate_pts <= BOUNDARY_MARGIN) | (candidate_pts >= 1.0 - BOUNDARY_MARGIN), axis=1) if BOUNDARY_MARGIN > 0 else np.zeros(len(candidate_pts), dtype=bool)
masked = too_close | near_boundary
EI_RBF[masked] = EI_Matern[masked] = EI_RBF_noise[masked] = -np.inf
PI_RBF[masked] = PI_Matern[masked] = PI_RBF_noise[masked] = -np.inf
LCB_RBF[masked] = LCB_Matern[masked] = LCB_RBF_noise[masked] = np.inf

def _pick(acq_val, maximize, fallback):
    idx = np.argmax(acq_val) if maximize else np.argmin(acq_val)
    return np.asarray(fallback).ravel() if masked[idx] else candidate_pts[idx]

x_best_EI_RBF  = _pick(EI_RBF, True, next_x_high_dist)
x_best_PI_RBF  = _pick(PI_RBF, True, next_x_high_dist)
x_best_UCB_RBF = _pick(LCB_RBF, False, next_x_high_dist)
x_best_EI_Matern  = _pick(EI_Matern, True, next_x_high_dist)
x_best_PI_Matern  = _pick(PI_Matern, True, next_x_high_dist)
x_best_UCB_Matern = _pick(LCB_Matern, False, next_x_high_dist)
x_best_EI_RBF_noise  = _pick(EI_RBF_noise, True, next_x_high_dist)
x_best_PI_RBF_noise  = _pick(PI_RBF_noise, True, next_x_high_dist)
x_best_UCB_RBF_noise = _pick(LCB_RBF_noise, False, next_x_high_dist)

x_best_EI  = x_best_EI_RBF  if is_rbf else (x_best_EI_Matern  if is_matern else x_best_EI_RBF_noise)
x_best_PI  = x_best_PI_RBF  if is_rbf else (x_best_PI_Matern  if is_matern else x_best_PI_RBF_noise)
x_best_UCB = x_best_UCB_RBF if is_rbf else (x_best_UCB_Matern if is_matern else x_best_UCB_RBF_noise)

print(f"Best kernel: {best_name}")
print(f"  EI  argmax: ({x_best_EI[0]:.4f}, {x_best_EI[1]:.4f}, {x_best_EI[2]:.4f}, {x_best_EI[3]:.4f})")
print(f"  PI  argmax: ({x_best_PI[0]:.4f}, {x_best_PI[1]:.4f}, {x_best_PI[2]:.4f}, {x_best_PI[3]:.4f})")
print(f"  UCB argmax: ({x_best_UCB[0]:.4f}, {x_best_UCB[1]:.4f}, {x_best_UCB[2]:.4f}, {x_best_UCB[3]:.4f})")
```

### 3.3 Acquisition surfaces

EI/PI/UCB are evaluated on `candidate_pts` (no 2D contour for 4D). **Section 5** shows the suggested next query and EI/PI/UCB argmax locations on GP mean/std slices.

### Pairwise distances between suggestions (small = agree)Small values mean two strategies agree; large values mean the next query depends on strategy choice.

```python
names = ['EI', 'UCB', 'PI']
pts = [np.asarray(x).ravel() for x in [x_best_EI, x_best_UCB, x_best_PI]]
n_pts = len(pts)
dists = np.zeros((n_pts, n_pts))
for i in range(n_pts):
    for j in range(n_pts):
        dists[i, j] = np.linalg.norm(pts[i] - pts[j])
for i in range(n_pts):
    for j in range(i + 1, n_pts):
        print(f"  {names[i]} vs {names[j]}: {dists[i, j]:.4f}")
max_d_3 = max(np.linalg.norm(pts[0]-pts[1]), np.linalg.norm(pts[0]-pts[2]), np.linalg.norm(pts[1]-pts[2]))
if max_d_3 < AGREE_THRESHOLD:
    print(f"\n>>> EI/PI/UCB agree (max pairwise dist = {max_d_3:.4f} < {AGREE_THRESHOLD}).")
else:
    print(f"\n>>> EI/PI/UCB differ (max dist = {max_d_3:.4f}); ensemble would use centroid.")
```

## 4. Select next query

- **Ensemble mode** (`USE_ENSEMBLE=True`): if EI, PI, UCB agree (pairwise distance &lt; `AGREE_THRESHOLD`) → use EI argmax; else → use centroid.
- **Solo mode** (`USE_ENSEMBLE=False`): use `SOLO_STRATEGY` (EI / PI / UCB).

See `docs_private/40_notes_and_references/ensemble_acquisition_guide.md`.

```python
x_EI  = np.asarray(x_best_EI).ravel()
x_PI  = np.asarray(x_best_PI).ravel()
x_UCB = np.asarray(x_best_UCB).ravel()
pts = np.array([x_EI, x_PI, x_UCB])
d12 = np.linalg.norm(pts[0] - pts[1])
d13 = np.linalg.norm(pts[0] - pts[2])
d23 = np.linalg.norm(pts[1] - pts[2])
max_d = max(d12, d13, d23)

_solo_map = {"EI": x_EI, "PI": x_PI, "UCB": x_UCB}
next_x_no_ens = np.clip(_solo_map.get(SOLO_STRATEGY, x_EI).copy(), 0.0, 0.999999)
next_x_with_ens = np.clip(np.mean(pts, axis=0), 0.0, 0.999999) if max_d >= AGREE_THRESHOLD else np.clip(x_EI.copy(), 0.0, 0.999999)

fmt = lambda v: ", ".join(f"{x:.4f}" for x in v)
print(f"Without ensemble ({SOLO_STRATEGY}):  x = ({fmt(next_x_no_ens)})")
print(f"With ensemble: EI=({fmt(x_EI)}), PI=({fmt(x_PI)}), UCB=({fmt(x_UCB)}) → max_dist={max_d:.4f}")
print(f"  → {'centroid' if max_d >= AGREE_THRESHOLD else 'all agree, EI'} = ({fmt(next_x_with_ens)})")

next_x = next_x_with_ens if USE_ENSEMBLE else next_x_no_ens

# Proximity check: only replace with fallback for near-duplicates.
DUPLICATE_THRESHOLD = 1e-3
dist_to_obs = np.sqrt(((X - next_x) ** 2).sum(axis=1))
nearest_idx  = np.argmin(dist_to_obs)
nearest_dist = dist_to_obs[nearest_idx]
if nearest_dist < DUPLICATE_THRESHOLD:
    print(f"⚠ Proposed point is duplicate (dist={nearest_dist:.2e} to obs #{nearest_idx+1}); using high-dist fallback.")
    next_x = next_x_high_dist
elif nearest_dist < MIN_DIST_THRESHOLD:
    print(f"  (Proposed point is {nearest_dist:.4f} from obs #{nearest_idx+1} — keeping it.)")

_strat = 'ensemble' if USE_ENSEMBLE else SOLO_STRATEGY
print(f"Next query | kernel: {best_name} | strategy: {_strat} | xi={XI_EI_PI}")
print(f"  x = ({fmt(next_x)})")
print(f"  Nearest existing obs: #{nearest_idx+1} at dist={nearest_dist:.4f}")
print(f"  Current best y = {best_y:.4g} at ({fmt(best_x_so_far)})")
```
